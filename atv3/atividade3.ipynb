```python
import pandas as pd
```

```python
# Leitura dos dados
df= pd.read_csv('Cópia de dados_atividade_3_alunos - dados_atividade_3.csv')
df
```

```python
# Tratamento dos dados

# Separa a coluna que contém os dados (coluna 1) em 4 colunas e remove as linhas inúteis
df = dict(df['Abastecimento de água - Brasil'].str.split(';')[2:-3])

# Faz a transposição 
df = pd.DataFrame(df).T

# Atualiza o header do df para a primeira linha e remove aspas das strings
df.columns = df.iloc[0].str.replace(pat="\"",repl="")

# Reseta o índice agora com um novo df
df.reset_index(drop=True, inplace=True)

# Remove a primeira linha agora que ela é o header
df = df.drop(0)

# Remove nome da coluna de índices
df.columns.name = None
df 
```

PRÁTICA 1: Utilizando a linguagem R ou Python obtenha para cada ano uma amostra aleatória simples de
70 municípios.

```python
import numpy as np
from scipy import stats
```

```python
df.columns
print(df['Município'].iloc[0])
```

```python
# Extrai amostras aleatórias simples de 70 municípios para cada ano
amostra = { 1991 : [], 2000 : [], 2010 : [] }
for ano in amostra:
    m = np.random.choice(df.index, 70)
    for idx in m:
        amostra[ano].append((df['Município'].iloc[idx],df[str(ano)].iloc[idx]))

# Imprime as amostras por ano
for ano in amostra:
    print(f"Amostras do ano {ano}:")
    for idx, a in enumerate(amostra[ano]):
        print(idx, ":",a)
```

PRÁTICA 3: Utilizando a linguagem R ou Python obtenha para cada ano as medidas de Centralidade, ou
seja, Média, Moda e Mediana

```python
'''
Função que retorna o valor inteiro de um elemento se o mesmo for diferente de - (hífen)
'''
def retorna_int(s):
    return int(s) if s != '-' else 0
```

```python
# Calcula a média dos anos
# OBS: Como os valores sem dados foram substituidos por 0 para poder calcular as métricas, as 
# modas dos anos 1991 e 2000 acabaram por ser 0
medias = {}
modas = {}
medianas = {}

# dict com os valores para não precisar recalcular abaixo
valores = {}

for ano in df.columns[1:]:
    valores[ano] = list(map(retorna_int,list(df[ano].values)))
    medias[ano] = np.mean(valores[ano])
    modas[ano] = int(stats.mode(valores[ano],keepdims=True)[0])
    medianas[ano] = np.median(valores[ano])

for ano in medias:
    print(f"Ano {ano} - Média: {medias[ano]:.2f} - Moda: {modas[ano]} - Mediana: {medianas[ano]}")
```

PRÁTICA 4: Utilizando a linguagem R ou Python obtenha para cada ano as medidas de Variabilidade, ou
seja, Variância, Desvio Padrão, Amplitude e os Quartis.

```python
vars = {}
dps = {}
ampls = {}
quartis = {}

for ano in df.columns[1:]:
    vars[ano] = np.var(valores[ano], ddof=0)
    dps[ano] = np.std(valores[ano])
    ampls[ano] = np.linalg.norm(valores[ano])
    quartis[ano] = { 'Q1': np.percentile(valores[ano], 25), 'Q3' : np.percentile(valores[ano],75)}
    
# print(f"Variancia: {vars}\n Desvio Padrao: {dps}\n Amplitude: {ampls}\n Quartis: {quartis}")

for ano in medias:
    print(f"Ano {ano} - Variância: {vars[ano]:.2f} - Desvio Padrão: {dps[ano]} - Amplitude: {ampls[ano]} - Quartis: {quartis[ano]}")

df
```
